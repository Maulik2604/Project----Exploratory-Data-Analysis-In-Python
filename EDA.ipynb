Exploratory Data Analysis In Python

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Dataset For EDA.csv")
df.head()
```

```python
df.dtypes
```

```python
df= df.drop(['Date of Birth'],axis=1)
df.head(10)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df=df.drop_duplicates()
df.head()
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
df=df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Year of Admission'])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Fees'], y=df['Discount on Fees'])
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,annot=True)
```
